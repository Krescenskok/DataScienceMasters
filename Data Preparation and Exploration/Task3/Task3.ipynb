# Encoding

```python
# Logic adapted from GeeksforGeeks (2025e)
# Source: https://www.geeksforgeeks.org/machine-learning/how-to-perform-ordinal-encoding-using-sklearn/

from sklearn.preprocessing import OrdinalEncoder

# select categories to encode
categories = [["Prefer not to answer", "Very Dissatisfied", "Dissatisfied", "Satisfied", "Very Satisfied"]]

encoder = OrdinalEncoder(categories=categories, dtype=int)
df_encoded = df.copy()
# encode values and assign to new column
df_encoded[["Satisfaction_Score"]] = encoder.fit_transform(df_encoded[["CustomerOrderSatisfaction"]])
# drop original column
df_encoded.drop(columns=['CustomerOrderSatisfaction'], inplace=True)
df_encoded['Satisfaction_Score'].value_counts()
```

```python
# select categories to encode
categories = [["Medium", "High"]]
encoder = OrdinalEncoder(categories=categories, dtype=int)
# encode values and assign to new column
df_encoded[["OrderPriority_Score"]] = encoder.fit_transform(df_encoded[["OrderPriority"]])
# drop original column
df_encoded.drop(columns=['OrderPriority'], inplace=True)
df_encoded["OrderPriority_Score"].value_counts()
```

```python
# encode the 2 columns
df_encoded = pd.get_dummies(df, columns=["DiscountApplied", "Segment"], drop_first=True, prefix=["DiscountApplied", "Segment"])
# Map values to 1 and 0
df_encoded['DiscountApplied_Yes'] = df_encoded['DiscountApplied_Yes'].map({True: 1, False: 0})
df_encoded['Segment_Corporate'] = df_encoded['Segment_Corporate'].map({True: 1, False: 0})
df_encoded.head()
```

### Encode All Categorical Variables

```python
# encode remaining columns
df_encoded = pd.get_dummies(df_encoded, columns=['Country', 'ExpeditedShipping', 'Region', 'PaymentMethod', 'ProductName'], 
                            prefix=['Country', 'ExpeditedShipping', 'Region', 'PaymentMethod', 'ProductName'])
# map values to 1 or 0
df_encoded = df_encoded.apply(lambda col: col.map({True: 1, False: 0}) if col.dtype == 'bool' else col)
```

### Transactionalize Data

```python
df.info()
```

```python
# only keep columns necessary for transactionalizing
cols_keep = ['OrderID', 'ProductName']
df = df[cols_keep]
df.head()
```

```python
# encode product names
df = pd.get_dummies(df, columns=['ProductName'], drop_first=True)
# group by the order ID to get products in a singular transaction
df = df.groupby('OrderID').any().reset_index()
df.drop(columns=['OrderID'], inplace= True)
df.head()
```

```python
# export df to csv
df.to_csv('Transactional.csv', index=False)
```

# Market Basket Analysis

```python
# Logic adapted from GeeksforGeeks (2025f)
# Source: https://www.geeksforgeeks.org/machine-learning/apriori-algorithm/
from mlxtend.frequent_patterns import apriori, association_rules

# use apriori function where the data exists at least 1% of the time
itemsets = apriori(df, min_support=.01, use_colnames=True)
print(itemsets.shape[0])
```

### Find Association Rules

```python
# find association rules based on lift
rules = association_rules(itemsets, metric="lift", min_threshold=1)
rules.shape[0]
print(rules.head())
```

```python
# sort rules by the lift value 
rules = rules.sort_values(by='lift', ascending=False)
pd.set_option("display.max_colwidth", None)
# print top 3 rules with metrics
rules[['antecedents', 'consequents', 'lift', 'confidence', 'support']].head(3)
```
