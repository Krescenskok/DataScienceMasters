### Parametric Testing
- ANOVA

```python
# import package
from scipy.stats import f_oneway


# split bmi data by region
se = df[df['region'] == 'southeast']['bmi']
sw = df[df['region'] == 'southwest']['bmi']
ne = df[df['region'] == 'northeast']['bmi']
nw = df[df['region'] == 'northwest']['bmi']
```

```python
# Logic adapted from GeeksforGeeks (2025d)
# Source: https://www.geeksforgeeks.org/python/how-to-perform-a-one-way-anova-in-python/

# perform anova test on bmi and regions
f_statistic, p_value = f_oneway(se, sw, ne, nw)
print(p_value)
```

```python
# print f statistic
print(f_statistic)
```

### Non-Parametric Testing

```python
# Ensure charges is a numeric data type before conducting analysis
df.info()
```

```python
# Create a histogram to understand the distribution (not normal)
sns.histplot(x="charges", data=df, kde=True, bins=10)
plt.title("Histogram of Charges")
plt.show()
```

```python
df.sex.value_counts()
```

```python
# Logic adapted from (Mannwhitneyu — SciPy v1.16.2 Manual, n.d.)
# Source: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html

from scipy.stats import mannwhitneyu

# split the dataset into charges for males and charges for females
males = df[df['sex'] == 'male']['charges']
females = df[df['sex'] == 'female']['charges']
# perform mann whitney u test
u1, p = mannwhitneyu(males, females)
# print u statistic
print(u1)
```

```python
# print p-value
print(p)
```
